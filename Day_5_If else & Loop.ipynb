```python
n = int(input("Enter a number: "))
rev = 0
while n != 0:
    digit = n % 10
    rev = rev * 10 + digit
    n //= 10
print("Reversed number:", rev)
```

```python
n = int(input("Enter a number: "))
digit_sum = 0
temp = n
while temp != 0:
    digit_sum += temp % 10
    temp //= 10
print("Sum of digits of", n, "=", digit_sum)
```

```python
n = int(input("Enter a number: "))
original = n
rev = 0
while n != 0:
    rev = rev * 10 + n % 10
    n //= 10
if original == rev:
    print(original, "is a palindrome")
else:
    print(original, "is not a palindrome")
```

```python
n = int(input("Enter a number: "))
fact = 1
i = 1
while i <= n:
    fact *= i
    i += 1
print("Factorial:", fact)
```
