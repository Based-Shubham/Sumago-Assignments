```python
# Task 28
def largest_in_list(lst):
    largest = lst[0]
    for num in lst:
        if num > largest:
            largest = num
    return largest
print(largest_in_list([45, 12, 78, 34, 90]))
```

```python
# Task 28
def get_max(items):
    max_val = items[0]
    for val in items:
        if val > max_val:
            max_val = val
    return max_val
print(get_max([88, 45, 92, 67, 55]))
```
