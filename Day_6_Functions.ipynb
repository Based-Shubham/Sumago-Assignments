```python
# Task 42
def is_palindrome_num(n):
    original = n
    rev = 0
    while n != 0:
        rev = rev * 10 + n % 10
        n //= 10
    if original == rev:
        print("Palindrome")
    else:
        print("Not Palindrome")
is_palindrome_num(121)
```

```python
# Task 42
def check_palindrome_num(num):
    orig = num
    reversed_num = 0
    while num != 0:
        reversed_num = reversed_num * 10 + num % 10
        num //= 10
    if orig == reversed_num:
        print("Palindrome")
    else:
        print("Not Palindrome")
check_palindrome_num(1234)
```

```python
# Task 45
def primes_upto_50():
    for n in range(2, 51):
        is_p = True
        for i in range(2, n):
            if n % i == 0:
                is_p = False
                break
        if is_p:
            print(n, end=" ")
primes_upto_50()
```

```python
# Task 45
def show_primes():
    for num in range(2, 51):
        prime = True
        for j in range(2, num):
            if num % j == 0:
                prime = False
                break
        if prime:
            print(num, end=" ")
show_primes()
```
