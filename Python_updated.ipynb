**Type Casting:**
The process of converting one data type into another.

- `int()` — converts to integer — `int(3.7)` → `3`
- `float()` — converts to float — `float(3)` → `3.0`
- `str()` — converts to string — `str(3)` → `'3'`
- `bool()` — converts to boolean — `bool(0)` → `False`, `bool(1)` → `True`

> ⚠️ When converting `float` to `int`, the decimal part is dropped, not rounded — `int(3.9)` → `3`

```python
a = 3
print(float(a))       # int  -> float  : 3.0

b = 3.7
print(int(b))         # float -> int   : 3  (decimal part dropped)

c = "123"
print(int(c))         # str  -> int    : 123
print(float(c))       # str  -> float  : 123.0

print(str(a))         # int  -> str    : '3'
print(type(str(a)))   # <class 'str'>
```

**Arithmetic Operators:**

- `*` — multiplication — `x * y`

**Assignment Operators:**
Used to assign values to variables.

**Identity Operators:**
Used to check whether two variables point to the same object in memory, not just the same value.

- `is` — returns `True` if both variables refer to the same object — `a is b` → `True`
- `is not` — returns `True` if both variables refer to different objects — `a is not c` → `True`

> 💡 `is` checks memory address (use `id()` to verify), whereas `==` checks value equality.

```python
a = [1, 2, 3]
b = a            # b points to the SAME object in memory
c = [1, 2, 3]    # c is a DIFFERENT object with the same values

print(a is b)     # True  — same memory address
print(a is c)     # False — different objects
print(a is not c) # True

# id() shows the memory address of the object
print(id(a), id(b))  # same number
print(id(c))         # different number
```

**Membership Operators:**
Used to check whether a value exists inside a sequence (string, list, tuple, set).

- `in` — returns `True` if the value is found in the sequence — `"a" in "Rasika"` → `True`
- `not in` — returns `True` if the value is not found in the sequence — `"z" not in "Rasika"` → `True`

```python
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)      # True
print("grape" in fruits)      # False
print("grape" not in fruits)  # True

# also works with strings
s = "Hello World"
print("Hello" in s)           # True
print("xyz" not in s)         # True
```

**List Comprehension:**
A concise way to create new lists by applying an expression to each item in an existing iterable — shorter and faster than a traditional `for` loop.

**General formula for a list comprehension:**

`new_list = [expression for item in iterable if condition]`

- `expression` — the operation or value to be included in the new list
- `item` — a placeholder that represents each element in the iterable
- `iterable` — any sequence to loop over (list, tuple, range, etc.)
- `if condition` — optional filter; only items where condition is True are included

```python
# write a program to print list of even numbers and list of odd numbers using list commprehension

even = [ x for x in range(0,101) if x%2 == 0]
print(even)
odd = [x for x in range(0,101) if x%2 != 0]
print(odd)
```

**Set Methods:**

- `.add()` — adds an item to the set — `s.add("hello")`
- `.update()` — adds elements from another list, tuple, or set — `s.update(s2)`
- `.remove()` — removes a specific element, raises error if not found — `s.remove("hello")`
- `.discard()` — removes an element, no error if not found (safer) — `s.discard("hello")`
- `.pop()` — removes and returns a random element — `s.pop()`
- `.clear()` — removes all elements from the set — `s.clear()`
- `del` — deletes the entire set variable from memory — `del s`
- `.union()` — returns a new set with all unique items from both sets — `s.union(s2)` or `s | s2`
- `.intersection()` — returns a new set with common items only — `s.intersection(s2)` or `s & s2`
- `.difference()` — returns items in first set but not in second — `s.difference(s2)` or `s - s2`

```python
s1 = {1, 2, 3}

# update() — adds all elements from another iterable into the set
s1.update([4, 5])    # add from a list
print(s1)            # {1, 2, 3, 4, 5}

s2 = {5, 6, 7}
s1.update(s2)        # merge another set (duplicates ignored)
print(s1)            # {1, 2, 3, 4, 5, 6, 7}
```

```python
s = {1, 2, 3, 4, 5}

# remove() — raises KeyError if element is not found
s.remove(3)
print(s)              # {1, 2, 4, 5}

# discard() — does NOT raise an error if element is missing (safer)
s.discard(10)         # 10 not in set — no error raised
print(s)              # {1, 2, 4, 5}

# del — deletes the entire set variable from memory
temp = {7, 8, 9}
del temp
# print(temp)  # would raise NameError: name 'temp' is not defined
```

**Set Operations (continued):**

- `.symmetric_difference()` — returns items in either set but **not in both** — `s.symmetric_difference(s2)` or `s ^ s2`
- `.issubset()` — returns `True` if all elements of the set are in another set — `s1.issubset(s2)`
- `.issuperset()` — returns `True` if the set contains all elements of another set — `s1.issuperset(s2)`
- `.isdisjoint()` — returns `True` if the two sets have **no common elements** — `s1.isdisjoint(s2)`


```python
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

# symmetric_difference() — items in either set, but NOT in both
print(s1.symmetric_difference(s2))  # {1, 2, 5, 6}
print(s1 ^ s2)                      # same using ^ operator
```

```python
a = {1, 2}
b = {1, 2, 3, 4}

# issubset() — all of a is inside b?
print(a.issubset(b))    # True
print(b.issubset(a))    # False

# issuperset() — b contains all of a?
print(b.issuperset(a))  # True
print(a.issuperset(b))  # False
```

```python
x = {1, 2, 3}
y = {4, 5, 6}
z = {2, 3, 4}

# isdisjoint() — True only if NO common elements
print(x.isdisjoint(y))  # True  — no overlap
print(x.isdisjoint(z))  # False — 2 and 3 are shared
```

**Frozen Set:**
A frozen set is an **immutable** version of a set — once created, elements cannot be added, removed, or changed.

- Created using `frozenset()` — `fs = frozenset({1, 2, 3})`
- Supports all read-only set operations — union, intersection, difference etc.
- Does **not** support `.add()`, `.remove()`, `.update()` — these will raise an `AttributeError`
- Can be used as a dictionary key (unlike a regular set) because it is hashable


```python
fs = frozenset({1, 2, 3, 4})
print(fs)           # frozenset({1, 2, 3, 4})
print(type(fs))     # <class 'frozenset'>

# Read-only operations still work
fs2 = frozenset({3, 4, 5})
print(fs | fs2)     # frozenset({1, 2, 3, 4, 5})  — union
print(fs & fs2)     # frozenset({3, 4})            — intersection

# This will raise AttributeError:
# fs.add(5)   — frozenset is immutable!
```

**Python Dictionary:**
A dictionary is an ordered collection of **key-value pairs**, where each key must be unique.

- Stores data as `key: value` pairs — `{"name": "Shubham", "age": 20}`
- Keys must be **unique** — duplicate keys overwrite the previous value
- Values **can** be duplicated
- Dictionaries are **mutable** — key-value pairs can be added, updated, or removed
- Keys must be **immutable** types — strings, integers, tuples (not lists)
- Dictionaries are **ordered** (from Python 3.7+) — insertion order is preserved
- Dictionaries are enclosed in curly brackets — `d = {}`
- An empty `{}` creates a **dictionary** (not a set — use `set()` for empty set)


```python
# Creating a dictionary
student = {"name": "Shubham", "age": 20, "city": "Nashik"}
print(student)

# Empty dictionary
empty = {}
print(type(empty))   # <class 'dict'>
```

**Accessing Dictionary Values:**

- Using key in square brackets — `d["key"]` — raises `KeyError` if key does not exist
- Using `.get()` — `d.get("key")` — returns `None` if key is missing (no error)
- Using `.get()` with a default — `d.get("key", default_value)` — returns default if key missing

> ⚠️ Dictionary does not support index-based access like lists — `d[0]` will cause a `KeyError` unless `0` is an actual key


```python
student = {"name": "Shubham", "age": 20, "city": "Nashik"}

# Access using key
print(student["name"])           # Shubham

# get() — safe access, no error if key missing
print(student.get("age"))        # 20
print(student.get("salary"))     # None — key not found, no crash
print(student.get("salary", 0))  # 0   — default value returned
```

**Adding and Updating Dictionary Values:**

- Add a new key-value pair — `d["new_key"] = value`
- Update an existing key — `d["existing_key"] = new_value` — just reassign it
- If the key already exists, the value is **overwritten**


```python
student = {"name": "Shubham", "age": 20}

# Add a new key-value pair
student["city"] = "Nashik"
print(student)   # {'name': 'Shubham', 'age': 20, 'city': 'Nashik'}

# Update an existing value
student["age"] = 21
print(student)   # {'name': 'Shubham', 'age': 21, 'city': 'Nashik'}
```

**Removing Items from a Dictionary:**

- `.pop(key)` — removes a specific key and **returns its value** — raises `KeyError` if missing
- `.popitem()` — removes and returns the **last inserted** key-value pair as a tuple
- `del d[key]` — deletes a specific key — raises `KeyError` if key does not exist
- `.clear()` — removes **all** key-value pairs — dictionary becomes empty `{}`

> 💡 `pop()` vs `popitem()` — `pop()` targets a specific key, `popitem()` always removes the last one


```python
d = {"a": 1, "b": 2, "c": 3}

# pop() — removes key and returns its value
val = d.pop("a")
print(val)   # 1
print(d)     # {'b': 2, 'c': 3}

# popitem() — removes last inserted pair, returns as tuple
item = d.popitem()
print(item)  # ('c', 3)
print(d)     # {'b': 2}

# del — deletes a specific key
d2 = {"x": 10, "y": 20}
del d2["x"]
print(d2)    # {'y': 20}

# clear() — empties the dictionary
d2.clear()
print(d2)    # {}
```

**Viewing Dictionary Data:**

- `.keys()` — returns all keys as a view object — `d.keys()`
- `.values()` — returns all values as a view object — `d.values()`
- `.items()` — returns all key-value pairs as a view of tuples — `d.items()`
- `len()` — returns the total number of key-value pairs — `len(d)`

> 💡 To get a plain list from these, wrap them in `list()` — `list(d.keys())`


```python
student = {"name": "Shubham", "age": 20, "city": "Nashik"}

print(student.keys())    # dict_keys(['name', 'age', 'city'])
print(student.values())  # dict_values(['Shubham', 20, 'Nashik'])
print(student.items())   # dict_items([('name','Shubham'),('age',20),('city','Nashik')])

print(len(student))      # 3 — number of key-value pairs

# Convert to list
print(list(student.keys()))  # ['name', 'age', 'city']
```

```python
student = {"name": "Shubham", "age": 20, "city": "Nashik"}

# Loop through keys only
for key in student:
    print(key, ":", student[key])

print()

# Loop through key-value pairs using items()
for key, value in student.items():
    print(key, "→", value)
```

**Checking Key Existence and Copying:**

- `in` operator — checks if a key exists — `"name" in d` → `True` or `False`
- `.copy()` — creates a **shallow copy** of the dictionary — `d2 = d.copy()`

> ⚠️ `d2 = d` does NOT create a copy — both variables point to the **same** dictionary in memory. Use `.copy()` to get an independent copy.


```python
student = {"name": "Shubham", "age": 20}

# Check if a key exists
print("name" in student)    # True
print("salary" in student)  # False

# copy() — independent copy
backup = student.copy()
backup["age"] = 99
print(student)  # {'name': 'Shubham', 'age': 20} — original unchanged
print(backup)   # {'name': 'Shubham', 'age': 99} — copy modified
```

**Nested Dictionary:**
A dictionary inside another dictionary.

- Used to represent complex or hierarchical data
- Access inner values by chaining keys — `d["outer_key"]["inner_key"]`


```python
student = {
    "name": "Shubham",
    "marks": {"math": 90, "science": 85, "english": 78}
}

print(student["name"])              # Shubham
print(student["marks"])             # {'math': 90, 'science': 85, 'english': 78}
print(student["marks"]["math"])    # 90
```

**Merging Two Dictionaries:**

- `.update(d2)` — adds all key-value pairs from `d2` into the calling dictionary — modifies in place
- If a key exists in both, the value from `d2` overwrites the original


```python
d1 = {"a": 1, "b": 2}
d2 = {"c": 3, "d": 4}

d1.update(d2)
print(d1)   # {'a': 1, 'b': 2, 'c': 3, 'd': 4}

# If keys overlap, d2 value wins
d3 = {"a": 1, "b": 2}
d4 = {"b": 99, "c": 3}
d3.update(d4)
print(d3)   # {'a': 1, 'b': 99, 'c': 3}
```

**Creating a Dictionary using `zip()`:**

- `zip()` pairs elements from two lists together
- Wrapping it with `dict()` converts those pairs into a dictionary
- Syntax: `dict(zip(keys_list, values_list))`


```python
keys = ["name", "age", "city"]
values = ["Shubham", 20, "Nashik"]

student = dict(zip(keys, values))
print(student)   # {'name': 'Shubham', 'age': 20, 'city': 'Nashik'}
```

**Sorting a Dictionary:**

- `sorted(d)` — returns a sorted **list of keys** in ascending order
- To sort by values — use `sorted(d, key=lambda x: d[x])`
- Dictionaries themselves cannot be sorted in-place; `sorted()` always returns a list


```python
marks = {"Math": 90, "English": 78, "Science": 85}

# Sort by keys (alphabetically)
sorted_keys = sorted(marks)
print(sorted_keys)    # ['English', 'Math', 'Science']

# Sort by values
sorted_by_val = sorted(marks, key=lambda x: marks[x])
print(sorted_by_val)  # ['English', 'Science', 'Math']  (ascending by marks)

# Find sum of all values
print(sum(marks.values()))     # 253

# Find key with minimum value (lowest scorer)
print(min(marks, key=lambda x: marks[x]))   # English

# Find key with maximum value (highest scorer)
print(max(marks, key=lambda x: marks[x]))   # Math
```

**Dictionary Comprehension:**
A concise way to create dictionaries — similar to list comprehension but uses `{}` and produces key-value pairs.

**General formula:**

`new_dict = {key_expression: value_expression for item in iterable if condition}`


```python
# Dictionary of squares from 1 to 10
squares = {x: x**2 for x in range(1, 11)}
print(squares)
# {1: 1, 2: 4, 3: 9, 4: 16, 5: 25, 6: 36, 7: 49, 8: 64, 9: 81, 10: 100}

# Dictionary of even squares only
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(even_squares)
# {2: 4, 4: 16, 6: 36, 8: 64, 10: 100}
```

**Dictionary Practical Uses:**

- Counting character frequency in a string
- Counting word frequency in a sentence
- Storing and finding the highest scorer


```python
# Count frequency of each character in a string
word = "python"
freq = {}
for ch in word:
    freq[ch] = freq.get(ch, 0) + 1
print(freq)   # {'p':1, 'y':1, 't':1, 'h':1, 'o':1, 'n':1}
```

```python
# Count word frequency in a sentence
sentence = "python is great and python is easy"
words = sentence.split()
word_count = {}
for w in words:
    word_count[w] = word_count.get(w, 0) + 1
print(word_count)
# {'python': 2, 'is': 2, 'great': 1, 'and': 1, 'easy': 1}
```

```python
# Store student marks and find the highest scorer
students = {"Shubham": 88, "Kaustubh": 92, "Rohan": 75, "Priya": 95}

# Find highest scorer
topper = max(students, key=lambda x: students[x])
print("Highest scorer:", topper, "with", students[topper], "marks")
# Highest scorer: Priya with 95 marks
```

**Python Functions:**
A function is a block of code used to perform repetitive tasks — it enhances code reusability and reduces redundancy.

- Defined using the `def` keyword — `def function_name(parameters):`
- Must be **called** to execute — `function_name()`
- Code inside a function runs only when the function is called
- Functions can optionally take **parameters** (inputs) and **return** a value
- Helps avoid writing the same code multiple times


**Types of Functions:**

1. **No argument, No return** — function takes no input and prints output directly
2. **With argument, No return** — function takes input values but prints output directly
3. **No argument, With return** — function takes no input but returns the result
4. **With argument, With return** — function takes input and returns the result

> 💡 `return` sends the result back to the caller so it can be stored or used further. Without `return`, the function just prints and the value is lost.


**Type 1 — No argument, No return:**
The function takes input inside itself and prints the result directly — no parameters, no `return`.


```python
# wap to find addition of two numbers using function
def addition():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    c = a + b
    print("The addition of two numbers is: ", c)

addition()  # calling the function
```

```python
# wap to find area of a circle using function
def area_of_circle():
    r = float(input("Enter the radius of the circle: "))
    area = 3.14 * r * r
    print("The area of the circle is: ", area)

area_of_circle()  # calling the function
```

```python
# wap to find if the number is even or odd using function
def even_odd():
    num = int(input("Enter a number: "))
    if num % 2 == 0:
        print(num, "is an even number.")
    else:
        print(num, "is an odd number.")

even_odd()  # calling the function
```

**Type 2 — With argument, No return:**
Values are passed to the function as parameters — the function uses them and prints directly.

- Parameters are defined inside `()` during function definition — `def sum(a, b):`
- Arguments are the actual values passed when calling — `sum(x, y)`


```python
# with argument no return
def sum(a, b):
    c = a + b
    print(c)

x = int(input("Enter 1st no: "))
y = int(input("Enter 2nd no: "))
sum(x, y)
```

```python
# wap to find volume of cube with argument no return
def area_cube(a):
    side = a
    formula = side ** 3
    print(formula)

x = int(input("Enter the side: "))
area_cube(x)
```

**Type 3 — No argument, With return:**
The function takes input inside itself but uses `return` to send the result back to the caller.

- `return` exits the function and passes the value back
- The returned value can be stored in a variable — `x = sum()`


```python
# no argument with return
def sum():
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    c = a + b
    return c

x = sum()
print("Addition:", x)
```

**Type 4 — With argument, With return:**
The cleanest and most reusable type — values passed in, result returned out.

- No input/output inside the function — fully controlled from outside
- Easy to test and reuse anywhere


```python
# with argument with return
def add(a, b):
    c = a + b
    return c  # final answer sent back

x = int(input("Enter a value: "))
y = int(input("Enter a value: "))
z = add(x, y)
print("Addition:", z)
```

**Default Parameters:**
A default value assigned to a parameter in the function definition — used when no argument is passed for that parameter.

- Syntax: `def function(a, b, c=5):` — `c` has a default value of `5`
- If the caller provides a value for `c`, the default is **overridden**
- If the caller skips `c`, the default value is **used automatically**
- Default parameters must come **after** non-default parameters

> ⚠️ You cannot place a non-default parameter after a default one — `def sum(a=1, b):` will cause a `SyntaxError`


```python
# without default — all 3 arguments required
def sum(a, b, c):
    d = a + b + c
    print(d)

sum(1, 3, 4)   # 8
```

```python
# with default — c has a default value of 5
def sum(a, b, c=5):
    d = a + b + c
    print(d)

sum(1, 2, 3)   # 6  — c = 3 (default overridden)
sum(1, 2)      # 8  — c = 5 (default used)
```

**The `pass` Statement:**
A placeholder that does nothing — prevents Python from throwing an error when a code block is left empty.

- Used when a function, loop, or condition is defined but code is not yet written
- Without `pass`, an empty block causes an `IndentationError` or `SyntaxError`
- Commonly used during development to plan structure before filling in logic


```python
def future():
    pass  # won't give an error — placeholder for future code

future()
```
