```python
basket_a = {"apple", "mango", "banana", "grape"}
basket_b = {"banana", "kiwi", "mango", "peach"}

while True:
    print("\n1. Union  2. Intersection  3. Difference  4. Symmetric Difference  5. Exit")
    choice = input("Choose an option: ")
    if choice == "1":
        print(basket_a | basket_b)
    elif choice == "2":
        print(basket_a & basket_b)
    elif choice == "3":
        print(basket_a - basket_b)
    elif choice == "4":
        print(basket_a ^ basket_b)
    elif choice == "5":
        print("Bye!")
        break
    else:
        print("Invalid choice")
```

```python
library = {"Python Basics": 5, "Data Science": 3, "AI Handbook": 7}

while True:
    print("\n1.View  2.Add  3.Update  4.Delete  5.Search  6.Exit")
    ch = input("Choice: ")
    if ch == "1":
        print(library)
    elif ch == "2":
        k = input("Book: "); v = int(input("Count: "))
        library[k] = v
    elif ch == "3":
        k = input("Book to update: "); v = int(input("New count: "))
        library[k] = v
    elif ch == "4":
        k = input("Book to delete: ")
        library.pop(k, None)
    elif ch == "5":
        k = input("Search book: ")
        print(library.get(k, "Not found"))
    elif ch == "6":
        break
```
